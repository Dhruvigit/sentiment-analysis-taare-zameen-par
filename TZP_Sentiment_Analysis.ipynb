```python
# --- You may need to install these libraries ---
# pip install pandas matplotlib seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# --- 1. Load and Prepare Data ---
df = pd.read_csv('Sentiment_Review_TZP.csv', encoding='utf-8')
# Clean column names by stripping leading/trailing whitespace
df.columns = df.columns.str.strip()

# --- Create sentiment column ---
rating_col = 'How would you rate Taare Zameen Par on a scale of 1 to 5?' # Define rating_col earlier
def assign_sentiment(rating):
    if rating is None: # Handle potential None values
        return None
    if rating >= 4:
        return 'Positive'
    elif rating == 3:
        return 'Neutral'
    else:
        return 'Negative'

df['sentiment'] = df[rating_col].apply(assign_sentiment)
# --- End of sentiment column creation ---

print("--- Starting Exploratory Data Analysis ---")
# Set plot style
sns.set_style("whitegrid")
plt.figure(figsize=(12, 8)) # Set a good default size

# --- 2. Analysis of Demographic Columns ---

# Age Group Distribution
age_col = 'What is your age group?'
plt.subplot(2, 2, 1) # Create a 2x2 grid, first plot
sns.countplot(data=df, x=age_col, palette='Blues_r', order=sorted(df[age_col].unique()))
plt.title('Distribution of Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Count')

# Gender Distribution
gender_col = 'What is your gender?'
plt.subplot(2, 2, 2) # Second plot
sns.countplot(data=df, x=gender_col, palette='Reds_r')
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')

# Rewatch Frequency
rewatch_col = 'How many times have you watched this movie?'
rewatch_order = ['Once', '2-3 times', 'More than 3 times']
plt.subplot(2, 2, 3) # Third plot
sns.countplot(data=df, x=rewatch_col, palette='Greens_r', order=rewatch_order)
plt.title('Rewatch Frequency')
plt.xlabel('')
plt.ylabel('Count')

# Save the demographic plots grid
plt.tight_layout() # Adjust layout to prevent overlap
plt.show()
print("Demographic analysis plot saved.")

# --- 3. Analysis of Rating-Based Columns ---
plt.figure(figsize=(10, 12)) # New figure for this set of plots

# Overall Movie Rating
rating_col = 'How would you rate Taare Zameen Par on a scale of 1 to 5?'
plt.subplot(3, 1, 1) # Create a 3x1 grid, first plot
sns.countplot(data=df, x=rating_col, palette='coolwarm')
plt.title('Overall Movie Rating Distribution')
plt.xlabel('Rating (1 to 5)')
plt.ylabel('Count')

# Dyslexia Portrayal Rating
dyslexia_col = 'How effectively did the movie portray the challenges faced by children with dyslexia?'
plt.subplot(3, 1, 2) # Second plot
sns.countplot(data=df, x=dyslexia_col, palette='viridis')
plt.title('Rating: Portrayal of Dyslexia')
plt.xlabel('Rating (1 to 5)')
plt.ylabel('Count')

# Education System Portrayal Rating
education_col = 'How do you feel about the way the film addressed the Indian education system and parental pressure?'
plt.subplot(3, 1, 3) # Third plot
sns.countplot(data=df, x=education_col, palette='plasma')
plt.title('Rating: Portrayal of Education System')
plt.xlabel('Rating (1 to 5)')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
print("Ratings analysis plot saved.")


# --- 4. Analysis of Categorical Response Columns ---
plt.figure(figsize=(10, 12)) # New figure

# Emotional Response
emotions_col = 'How did the movie make you feel?'
df[emotions_col] = df[emotions_col].fillna('')
all_emotions = [e.strip() for sublist in df[emotions_col].str.split(',') for e in sublist if e]
emotion_counts = Counter(all_emotions)
emotion_df = pd.DataFrame(emotion_counts.items(), columns=['Emotion', 'Count']).sort_values('Count', ascending=False)

plt.subplot(2, 1, 1) # Create a 2x1 grid, first plot
sns.barplot(data=emotion_df, x='Count', y='Emotion', palette='magma')
plt.title('Primary Emotions Evoked by the Movie')

# Favorite Character
character_col = 'Who was your favorite character?'
df['favorite_char_cleaned'] = df[character_col].str.split(',').str[0].str.strip().str.title()
df['favorite_char_cleaned'] = df['favorite_char_cleaned'].replace({
    'Ram': 'Ram Shankar Nikumbh', 'Nikumbh Sir': 'Ram Shankar Nikumbh', 'Aamir Khan': 'Ram Shankar Nikumbh',
    'Ishaan Awasthi': 'Ishaan', 'Ishan': 'Ishaan'
})

plt.subplot(2, 1, 2) # Second plot
sns.countplot(data=df, y='favorite_char_cleaned', order=df['favorite_char_cleaned'].value_counts().index, palette='crest')
plt.title('Favorite Character Mentions')
plt.xlabel('Number of Mentions')
plt.ylabel('Character')

plt.tight_layout()
plt.show()
print("Categorical responses plot saved.")
print("\nExploratory Data Analysis Complete.")
```

```python
# --- 2. Visualize Sentiment vs. Demographics ---
plt.figure(figsize=(12, 10))
sns.set_style("ticks")

# Age Group vs. Sentiment
age_col = 'What is your age group?'
plt.subplot(2, 1, 1)
sns.countplot(data=df, x=age_col, hue='sentiment', order=sorted(df[age_col].unique()),
              hue_order=['Positive', 'Neutral', 'Negative'], palette='viridis')
plt.title('Sentiment Distribution by Age Group')

# Gender vs. Sentiment
gender_col = 'What is your gender?'
plt.subplot(2, 1, 2)
sns.countplot(data=df, x=gender_col, hue='sentiment',
              hue_order=['Positive', 'Neutral', 'Negative'], palette='plasma')
plt.title('Sentiment Distribution by Gender')

plt.tight_layout()
plt.show()
print("Sentiment vs. Demographics plot saved.")


# --- 3. Visualize Sentiment vs. Ratings and Rewatches ---
plt.figure(figsize=(10, 12))

# Dyslexia Portrayal Rating vs. Sentiment
dyslexia_col = 'How effectively did the movie portray the challenges faced by children with dyslexia?'
plt.subplot(2, 1, 1)
sns.boxplot(data=df, x='sentiment', y=dyslexia_col, order=['Positive', 'Neutral', 'Negative'], palette='coolwarm')
plt.title('Rating for Dyslexia Portrayal by Sentiment Group')

# Rewatch Frequency vs. Sentiment
rewatch_col = 'How many times have you watched this movie?'
rewatch_order = ['Once', '2-3 times', 'More than 3 times']
plt.subplot(2, 1, 2)
sns.countplot(data=df, x=rewatch_col, hue='sentiment', order=rewatch_order,
              hue_order=['Positive', 'Neutral', 'Negative'], palette='magma')
plt.title('Sentiment by Rewatch Frequency')

plt.tight_layout()
plt.show()
print("Sentiment vs. Ratings & Rewatch plot saved.")


# --- 4. Visualize Sentiment vs. Categorical Responses ---
plt.figure(figsize=(12, 10))

# Favorite Character vs. Sentiment
character_col = 'Who was your favorite character?'
df['favorite_char_cleaned'] = df[character_col].str.split(',').str[0].str.strip().str.title()
df['favorite_char_cleaned'] = df['favorite_char_cleaned'].replace({
    'Ram': 'Ram Shankar Nikumbh', 'Nikumbh Sir': 'Ram Shankar Nikumbh', 'Aamir Khan': 'Ram Shankar Nikumbh',
    'Ishaan Awasthi': 'Ishaan', 'Ishan': 'Ishaan'
})
top_characters = df['favorite_char_cleaned'].value_counts().nlargest(3).index
df_top_chars = df[df['favorite_char_cleaned'].isin(top_characters)]

plt.subplot(2, 1, 1)
sns.countplot(data=df_top_chars, x='favorite_char_cleaned', hue='sentiment',
              order=top_characters, hue_order=['Positive', 'Neutral', 'Negative'], palette='Set2')
plt.title('Overall Sentiment of People Who Chose These Favorite Characters')

# Top Emotions within each Sentiment
emotions_col = 'How did the movie make you feel?'
df[emotions_col] = df[emotions_col].fillna('')
df['emotions_list'] = df[emotions_col].apply(lambda x: [e.strip() for e in x.split(',') if e])

positive_emotions = Counter([item for sublist in df[df['sentiment'] == 'Positive']['emotions_list'] for item in sublist])
negative_emotions = Counter([item for sublist in df[df['sentiment'] == 'Negative']['emotions_list'] for item in sublist])

df_pos_emo = pd.DataFrame(positive_emotions.most_common(5), columns=['Emotion', 'Count'])
df_pos_emo['Sentiment'] = 'Positive'
df_neg_emo = pd.DataFrame(negative_emotions.most_common(5), columns=['Emotion', 'Count'])
df_neg_emo['Sentiment'] = 'Negative'
df_emo_combined = pd.concat([df_pos_emo, df_neg_emo])

plt.subplot(2, 1, 2)
sns.barplot(data=df_emo_combined, x='Count', y='Emotion', hue='Sentiment', palette='Paired')
plt.title('Top Emotions Mentioned in Positive vs. Negative Reviews')

plt.tight_layout()
plt.show()
print("Sentiment vs. Categorical Responses plot saved.")
print("\nFinal integrated analysis complete.")
```

```python
import pandas as pd
import re
import emoji
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
import numpy as np # Import numpy here as it is used later in the cell


# 2. Prepare Data and Labels (sentiment column created in the previous cell)
review_col = 'Share your overall thoughts or review of the movie in your own words.'

df[review_col] = df[review_col].fillna('')

# 3. Preprocess Text (including emojis)
def preprocess_text(text):
    # Convert emojis to their text descriptions (e.g., 'ðŸ‘' -> ':thumbs_up:')
    text = emoji.demojize(str(text))
    # Remove punctuation and numbers, keeping the emoji text
    text = re.sub(r'[^a-z_]', ' ', text.lower())
    return text

# Apply the preprocessing to the review column
df['processed_review'] = df[review_col].apply(preprocess_text)

# 4. Convert Text to Numbers (Vectorization)
vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['processed_review'])
y = df['sentiment']

# 5. Split Data, Train Model, and Evaluate
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=3)

model = MultinomialNB()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

# --- Display Results ---
print(f"Model Accuracy: {accuracy:.2f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred, zero_division=0))

print("-x-x-x-x-x-x-x--x-x-x-x-x-x-x-x-x--x-x-x-x-x--x-x-x-x-x-x-")
from sklearn.feature_extraction.text import TfidfVectorizer
# 4. Vectorize Text using TF-IDF and N-grams (Strategy 1)
vectorizer2 = TfidfVectorizer(stop_words='english', ngram_range=(1, 2))
X2 = vectorizer2.fit_transform(df['processed_review'])
y2 = df['sentiment']

# 5. Split Data, Train Model, and Evaluate
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.25, random_state=1)

model2 = MultinomialNB()
model2.fit(X_train2, y_train2)

y_pred2 = model2.predict(X_test2)
accuracy2 = accuracy_score(y_test2, y_pred2)

# --- Display Results ---
print(f"Model Accuracy: {accuracy2:.2f}")
print("\nClassification Report:")
print(classification_report(y_test2, y_pred2, zero_division=0))

print("-x-x-x-x--x-x-x-x-x-x--x-x-x-x-x-x-x--x-x-x--xx")
from sklearn.preprocessing import MinMaxScaler
# --- BERT-specific imports ---
from transformers import BertTokenizer, BertModel
import torch
# 3. Initialize Pre-trained BERT Model and Tokenizer
print("Loading BERT model and tokenizer...")
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model3 = BertModel.from_pretrained('bert-base-uncased')
print("BERT model loaded.")

# 4. Generate BERT Embeddings
def get_bert_embedding(text, model3, tokenizer):
    # Tokenize the text and get PyTorch tensors
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=128)

    # Get model output without calculating gradients since we are not training bert but using for prediction
    with torch.no_grad():
        outputs = model3(**inputs)

    # Use the embedding of the [CLS] token, which represents the entire sequence's meaning to get output
    cls_embedding = outputs.last_hidden_state[:,0,:].numpy().flatten()
    return cls_embedding

print("\nGenerating BERT embeddings for all reviews... (This may take a moment)")
X_bert = np.array([get_bert_embedding(review, model3, tokenizer) for review in df['processed_review']])
y3 = df['sentiment']
print("Embeddings generated.")

# 5. Scale Features and Split Data
# MultinomialNB requires non-negative features, so we scale the BERT vectors to a [0, 1] range
scaler = MinMaxScaler()
X_bert_scaled = scaler.fit_transform(X_bert)

X_train3, X_test3, y_train3, y_test3 = train_test_split(X_bert_scaled, y3, test_size=0.3, random_state=42, stratify=y3)

# 6. Train and Evaluate the Naive Bayes Model
model_nb_on_bert = MultinomialNB()
model_nb_on_bert.fit(X_train3, y_train3)

y_pred3 = model_nb_on_bert.predict(X_test3)
accuracy3 = accuracy_score(y_test3, y_pred3)

# --- 7. Display Results ---
print("\n--- Naive Bayes Model with BERT Embeddings ---")
print(f"Model Accuracy: {accuracy3:.2f}")
print("\nClassification Report:")
print(classification_report(y_test3, y_pred3, zero_division=0))
```
